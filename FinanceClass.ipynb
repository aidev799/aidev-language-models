```python

# openai.organization = "org-0aeOV8xLMdS6EktEfObsesQ9"
# openai.api_key = os.getenv("OPENAI_API_KEY")

import requests

# Define your API key and endpoint
api_key = openai.api_key
endpoint = "https://api.openai.com/v1/completions"  # Use the appropriate endpoint for your API call

# Define your prompt or input text

# Set up the headers with your API key
headers = {
    "Authorization": f"Bearer {openai.api_key}",
    "Content-Type": "application/json",
}
Chunk=16
TransactionDescriptionList= str(H1Transactions['Description'].tolist()[(Chunk-1)*50:Chunk*50])
Categories=str(H1Transactions['Category'].unique())
# prompt = "Classify the following purchase description into one of the following categories in this list:(lunch, coffee, clothes, baby). The purchase descriptionis:('3CPAYMENT*STARBUCKS 127)"
prompt = ("I will provide you an ordered list of transaction descriptions. I would like you to return a list of values as a string to populate a new column in my table called Category, with an ordered list of Categories in the same order as the original list.  Classify the following purchase descriptions into one of the following categories in this list:"+Categories+" The purchase descriptions is: " + TransactionDescriptionList)
# Define the data for your API request
data = {
    "model": "gpt-3.5-turbo-instruct",
    "prompt": prompt,
    "max_tokens": 1500,
    "temperature": 0
}

# Make the API request
response = requests.post(endpoint, json=data, headers=headers)

# Check for a successful response and process the output
if response.status_code == 200:
    result = response.json()
    generated_text = result["choices"][0]["text"]
    print(generated_text)
else:
    print(f"Error: {response.status_code} - {response.text}")
```
